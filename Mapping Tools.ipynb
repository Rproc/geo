```python
import geopandas
import fiona
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
cities = geopandas.read_file(geopandas.datasets.get_path('naturalearth_cities'))
```

```python
world = world[(world.pop_est>0) & (world.name!="Antarctica")]
```

```python
world['gdp_per_cap'] = world.gdp_md_est / world.pop_est
```

```python
world.plot(column='gdp_per_cap', figsize=(15, 10))
```

```python
fig, ax = plt.subplots(1, 1)
world.plot(column='pop_est', ax=ax, legend=True, figsize=(15, 10))
```

```python
fig, ax = plt.subplots(1, 1)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
world.plot(column='pop_est', ax=ax, legend=True, cax=cax, figsize=(10, 7))
```

```python
fig, ax = plt.subplots(1, 1)

world.plot(column='pop_est', ax=ax, legend=True,legend_kwds={'label': "Population by Country", 'orientation': "horizontal"}, figsize=(20, 15))
```
